```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
pd.set_option('display.max_rows',None)
%matplotlib inline
```

```python
subscription_data=pd.read_csv(r'G:\MLDPI\Data analyst projects\Entri app Challenge\user_data_set\track_subscription_data.csv')
```

```python
purchase_data=pd.read_csv(r'G:\MLDPI\Data analyst projects\Entri app Challenge\user_data_set\user_purchase_data.csv')
```

```python
video_activity_data=pd.read_csv(r'G:\MLDPI\Data analyst projects\Entri app Challenge\user_data_set\user_video_activity_data.csv')
```

```python
quiz_attempt_data=pd.read_csv(r'G:\MLDPI\Data analyst projects\Entri app Challenge\user_data_set\user_quiz_attempt_data.csv')
```

```python
studycard_data=pd.read_csv(r'G:\MLDPI\Data analyst projects\Entri app Challenge\user_data_set\user_studycard_data.csv')
```

```python
quiz_attempt_data.head()
```

```python
# here in quiz_attempt_data replace the string None in the started_at column to NaN value
quiz_attempt_data['started_at']=np.where(quiz_attempt_data['started_at']=='None',np.nan,quiz_attempt_data['started_at'])
```

```python
purchase_data['date_joined']=pd.to_datetime(purchase_data['date_joined'],infer_datetime_format=True)
```

```python
purchase_data['date_of_first_purhcase']=pd.to_datetime(purchase_data['date_of_first_purhcase'],infer_datetime_format=True)
```

```python
subscription_data['subscription_date']=pd.to_datetime(subscription_data['subscription_date'],infer_datetime_format=True)
```

```python
video_activity_data['watched_at']=pd.to_datetime(video_activity_data['watched_at'],infer_datetime_format=True)
```

```python
quiz_attempt_data['started_at']=pd.to_datetime(quiz_attempt_data['started_at'],infer_datetime_format=True)
```

```python
studycard_data['start_time']=pd.to_datetime(studycard_data['start_time'],infer_datetime_format=True)
```

```python
# In purchased data make a seperate columns  which contain information on whether the user is paid or unpaid

dummy_data=['Unpaid' if val is pd.NaT else 'Paid' for val in purchase_data['date_of_first_purhcase'] ]

purchase_data['Subscription'] = dummy_data
```

```python
purchase_data.head(50)
```

```python
purchase_data=purchase_data.rename(columns={'user_uid':'user_id'})
```

```python
# next we merge or innerjoin the purchased data to other dataset with respect to the user_uid column
subscription_data_merged=pd.merge(subscription_data, purchase_data, how ='inner', on ='user_id')
video_activity_data_merged=pd.merge(video_activity_data, purchase_data, how ='inner', on ='user_id')
quiz_attempt_data_merged=pd.merge(quiz_attempt_data, purchase_data, how ='inner', on ='user_id')
studycard_data_merged=pd.merge(studycard_data, purchase_data, how ='inner', on ='user_id')
```

```python
studycard_data_merged.info()
```

```python
video_activity_data_merged.head()
```

```python
purchase_data.head()
```

```python
purchase_data['diff_dates']=purchase_data['date_of_first_purhcase']-purchase_data['date_joined']
```

```python
purchase_data.head()
```

```python
purchase_data['diff_dates']=purchase_data['diff_dates'].dt.days
```

```python
purchase_data.shape
```

```python
purchase_data.info()
```

```python

plt.figure(figsize=(20,20))
purchase_data.groupby('date_joined')['diff_dates'].sum().plot(kind='barh')
```

```python
subscription_data_merged.head()
```

```python
# find out which courses have maximum conversion days
purchase_data_convert=purchase_data[(~ purchase_data['diff_dates'].isnull()) & (purchase_data_convert['diff_dates']!=0)]

subscription_data_samp=pd.merge(subscription_data, purchase_data_convert, how ='inner', on ='user_id')
```

```python
subscription_data_samp.groupby('track_slug')['diff_dates'].max().sort_values(ascending=False)[0:1]
```

```python
#plot a bar chart showing the maximum number of days required to convert to paid subscription for the courses given
plt.figure(figsize=(20,20))
subscription_data_samp.groupby('track_slug')['diff_dates'].max().plot(kind='barh')
plt.xlabel('Course Category')
plt.ylabel('Maximum days')
plt.title('Max days to convert')
```
