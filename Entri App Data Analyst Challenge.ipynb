### Entri: Data Analyst Interview Challenge Dataset
Along with this document, you’ll get a zip file with the following Dataset.
● User Details 1000 random users from a fixed time-period
    ○ User UID, Date Joined, Date of First Purchase (if they have made a
    purchase)
    
● Track Subscriptions - Track subscriptions of the 1000 users in the previous data set.
    ○ User UID, Track Slug (Track = Course category)
    
● Quiz Attempts - Quiz Attempts of the 1000 users in the first data set.
    ○ User UID, Quiz Attempt UID, Start Datetime, Status, Total Number of
    questions in the quiz, Number of questions attempted
    
● Watched Video - Video Watch history of the 1000 users in the first data set
    ○ User UID, Activity UID, Start Datetime, Duration of the Video, Watched
    Duration
        
● Interacted Study Cards - Study Cards interacted by the 1000 users in the first dataset
    ○ User UID, Activity UID, Start Datetime

## 1. Paid vs Unpaid - Difference in behaviour in the first week.

```python
# In purchased data make a seperate columns  which contain information on whether the user is paid or unpaid

dummy_data=['Unpaid' if val is pd.NaT else 'Paid' for val in purchase_data['date_of_first_purhcase'] ]

purchase_data['Subscription'] = dummy_data
```

```python
purchase_data.head(5)
```

```python
purchase_data=purchase_data.rename(columns={'user_uid':'user_id'})
```

```python
# next we merge or innerjoin the purchased data to other dataset with respect to the user_uid column
subscription_data_merged=pd.merge(subscription_data, purchase_data, how ='inner', on ='user_id')
video_activity_data_merged=pd.merge(video_activity_data, purchase_data, how ='inner', on ='user_id')
quiz_attempt_data_merged=pd.merge(quiz_attempt_data, purchase_data, how ='inner', on ='user_id')
studycard_data_merged=pd.merge(studycard_data, purchase_data, how ='inner', on ='user_id')
```

```python
# creating a list of seven dates that corresponds to week so that we can filter out and only pick the rows that 
# are necessary from each table.
datelist = pd.date_range(purchase_data['date_joined'][0].strftime("%Y-%m-%d"), periods=7).tolist()
```

```python
datelist
```

```python
# data 
video_activity_data_merged['New_date']=pd.to_datetime(video_activity_data_merged['watched_at'].dt.date)
video_activity_weekly=video_activity_data_merged[video_activity_data_merged['New_date'].isin(datelist)]

quiz_attempt_data_merged['started_at']=quiz_attempt_data_merged['started_at'].bfill(axis ='rows')
quiz_attempt_data_merged['New_date']=pd.to_datetime(quiz_attempt_data_merged['started_at'].dt.date)
quiz_attempt_weekly=quiz_attempt_data_merged[quiz_attempt_data_merged['New_date'].isin(datelist)]

studycard_data_merged['New_date']=pd.to_datetime(studycard_data_merged['start_time'].dt.date)
studycard_data_weekly=studycard_data_merged[studycard_data_merged['New_date'].isin(datelist)]
```

## 2. Per-user targets we should aim to increase conversion percentage

```python
# In unpaid sucscribers find out how many users enrolled in the different courses
plt.figure(figsize=(20,20))
ax=subscription_data_merged[subscription_data_merged['Subscription']=='Unpaid'].groupby('track_slug')['user_id'].count().plot(kind='barh')
for p in ax.patches:
    ax.annotate("%.2f" % p.get_width(), (p.get_x() + p.get_width(), p.get_y()), xytext=(5, 10), textcoords='offset points')
plt.xlabel('Course count')
plt.title('Course subscribed by each unpaid users')
```

#### Maximum number of unpaid users are enrolled in spoken english class

```python
# Then we just analyse number of days that took to convert from unpaid to paid subsriber
# below we just find the difference between joining date and purchase date
purchase_data['diff_dates']=purchase_data['date_of_first_purhcase']-purchase_data['date_joined']
purchase_data['diff_dates']=purchase_data['diff_dates'].dt.days

# find out which courses have maximum conversion days
purchase_data_convert=purchase_data[(~ purchase_data['diff_dates'].isnull()) & (purchase_data['diff_dates']!=0)]
subscription_data_samp=pd.merge(subscription_data, purchase_data_convert, how ='inner', on ='user_id')
subscription_data_samp.groupby('track_slug')['diff_dates'].max().sort_values(ascending=False)[0:1]
```

```python
#plot a bar chart showing the maximum number of days required to convert to paid subscription for the courses given
plt.figure(figsize=(20,25))
ax=subscription_data_samp.groupby('track_slug')['diff_dates'].max().plot(kind='barh',color='r')
for p in ax.patches:
    ax.annotate("%.2f" % p.get_width(), (p.get_x() + p.get_width(), p.get_y()), xytext=(5, 10), textcoords='offset points')
plt.xlabel('Course Category')
plt.ylabel('Maximum days')
plt.title('Max days to convert')
```
